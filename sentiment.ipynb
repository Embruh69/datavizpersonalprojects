```python
def get_reviews(appid, params={'json':1}):
        url = 'https://store.steampowered.com/appreviews/'
        response = requests.get(url=url+appid, params=params, 
        headers={'User-Agent': 'Mozilla/5.0'})
        return response.json()
```

```python
def get_n_reviews(appid, n, days):
    reviews = []
    cursor = '*'
    params = {
            'json' : 1,
            'filter' : 'all',
            'language' : 'english',
            'day_range' : days,
            'review_type' : 'all',
            'purchase_type' : 'all'
            }

    while n > 0:
        params['cursor'] = cursor.encode()
        params['num_per_page'] = min(100, n)
        n -= 100

        response = get_reviews(appid, params)
        cursor = response['cursor']
        reviews += response['reviews']

        if len(response['reviews']) < 100: break

    return reviews
```

```python
df['raw_score'] = df['Text'].apply(lambda Text: nltkvd.polarity_scores(Text))
df
```

```python
df['polarity'] = df['raw_score'].apply(lambda score_dict: score_dict['compound'])
df
```

```python
df['sentiment'] = df['polarity'].apply(lambda c: 'Positive' if c > 0 
                                                   else 'Negative' if c < 0 
                                                   else 'Neutral')
df
```
