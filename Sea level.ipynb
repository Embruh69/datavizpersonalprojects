```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
def draw_plot():
    # Read data from file
    df = pd.read_csv(r'C:\Users\user\data\epa-sea-level.csv')

    # Create scatter plot
    x = df['Year'].to_numpy()
    y = df['CSIRO Adjusted Sea Level'].to_numpy()
    plt.scatter(x, y, c='b', s=10)
    

    # Create first line of best fit
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    predicted_level = slope * 2050 + intercept

    # Create second line of best fit
    x_2000 = x[x >= 2000]
    y_2000 = y[x >= 2000]
    slope_2000, intercept_2000, r_value, p_value, std_err = linregress(x_2000, y_2000)
    predicted_level = slope * 2050 + intercept
    plt.scatter(x, y)
    plt.plot(x, slope * x + intercept, c='r')
    plt.plot([2000, 2100], [slope_2000 * 2000 + intercept_2000, slope_2000 * 2100 + 
                            intercept_2000], c='g')
    

    # Add labels and title
    plt.title('Rise in Sea Level')
    plt.xlabel('Year')
    plt.ylabel('Sea Level (inches)')
    
    plt.savefig('sea_level_plot.png')
    return plt.gca()
```

```python
draw_plot()
```
