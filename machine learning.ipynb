```python
import pandas as pd
import snscrape.modules.twitter as sc
```

```python
tweets = []

for i, tweet in enumerate(sc.TwitterSearchScraper('Battlefield 2042 since:2021-11-20 until:2022-01-01').get_items()):
    if i > 100000:
        break
        
    tweets.append([tweet.date, tweet.rawContent, tweet.user.username])
```

```python
df = pd.DataFrame(tweets, columns=['Datetime', 'Text', 'Username'])
df
```
